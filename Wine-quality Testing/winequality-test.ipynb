```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Compute average rating per region
region_ratings = df_cleaned.groupby('region')['rating'].mean().reset_index()

# **Exclude the specific value**
filtered_regions = region_ratings[region_ratings['region'] != "Vino de Espana"]

# **Top 10 Regions**
top_regions = filtered_regions.sort_values(by='rating', ascending=False).head(10)
heatmap_top = top_regions.pivot_table(values='rating', index='region')

# **Bottom 10 Regions**
bottom_regions = filtered_regions.sort_values(by='rating', ascending=True).head(10)
heatmap_bottom = bottom_regions.pivot_table(values='rating', index='region')

# **Plot Heatmap - Top 10 Regions**
plt.figure(figsize=(8, 12))
sns.heatmap(
    heatmap_top.sort_values(by='rating', ascending=False),
    annot=True,
    cmap='coolwarm',
    fmt=".2f",
    linewidths=0.5,
    vmin=3, vmax=5  # Keep consistent scale
)
plt.title("Top 10 Wine Regions by Average Rating")
plt.xlabel("Regions")
plt.ylabel("Average Rating")
plt.yticks(rotation=0)  # Ensure horizontal labels
plt.show()

# **Plot Heatmap - Bottom 10 Regions**
plt.figure(figsize=(8, 12))
sns.heatmap(
    heatmap_bottom.sort_values(by='rating', ascending=True),
    annot=True,
    cmap='coolwarm',
    fmt=".2f",
    linewidths=0.5,
    vmin=3, vmax=5  # Keep consistent scale
)
plt.title("Bottom 10 Wine Regions by Average Rating")
plt.xlabel("Regions")
plt.ylabel("Average Rating")
plt.yticks(rotation=0)  # Ensure horizontal labels
plt.show()
```

```python
import pandas as pd

# Define categories
red_wines = [
    "Toro Red", "Tempranillo", "Ribera Del Duero Red", "Priorat Red",
    "Rioja Red", "Grenache", "Syrah", "Monastrell", "Mencia",
    "Montsant Red", "Cabernet Sauvignon"
]

white_wines = [
    "Rioja White", "Verdejo", "Albarino", "Chardonnay", "Sauvignon Blanc"
]

cava_wines = ["Cava", "Sparkling"]

rose_wines = ["Rosé"]  # Add any rosé wines if available

# Apply categorization
df_cleaned["wine_category"] = df_cleaned["type"].apply(
    lambda x: "Red" if x in red_wines else
              "White" if x in white_wines else
              "Cava" if x in cava_wines else
              "Rosé" if x in rose_wines else "Other"
)

# Display grouped data
print(df_cleaned[['wine', 'type', 'wine_category']].drop_duplicates())

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Compute global average rating
global_avg = df_cleaned['rating'].mean()
smoothing_factor = 50  # Adjust this for better balance

# Compute weighted score using Bayesian averaging
df_cleaned['weighted_score'] = ((df_cleaned['rating'] * df_cleaned['num_reviews']) + (global_avg * smoothing_factor)) / (df_cleaned['num_reviews'] + smoothing_factor)

# Group by region and compute weighted average
region_scores = df_cleaned.groupby('region')['weighted_score'].mean().reset_index()
region_scores = region_scores.sort_values(by='weighted_score', ascending=False).head(20)  # Top 20 regions

# Pivot for heatmap format
heatmap_data = region_scores.pivot_table(values='weighted_score', index='region')

# Plot heatmap with highest-ranked regions at the top
plt.figure(figsize=(8, 12))
sns.heatmap(heatmap_data.sort_values(by='weighted_score', ascending=False), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Labels & Title
plt.title("Top 20 Preferred Wine Regions (Adjusted for Ratings & Reviews)")
plt.xlabel("Regions")
plt.ylabel("Weighted Score")
plt.show()



```


Instead of a simple multiplication, we have used the Bayesian Average or Weighted Mean approach to adjust ratings based on review count. 

### Weighted Score Formula

The weighted score is calculated as follows:

\[
Weighted\_Score = \frac{{(avg\_rating \times num\_reviews) + (global\_avg \times smoothing\_factor)}}{{num\_reviews + smoothing\_factor}}
\]


where:
- **`avg_rating`** = average rating of the region
- **`num_reviews`** = total number of reviews for that region
- **`global_avg`** = overall average rating across all regions
- **`smoothing_factor`** = adjustment factor to balance scores for regions with low review counts (e.g., **50**)

This formula ensures that ratings with **more reviews carry more influence**, while preventing biases towards regions with very few reviews.
