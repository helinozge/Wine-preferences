```python
# Define custom price bins and labels
price_bins = [4.9, 50, 200, df_cleaned['price'].max()]
price_labels = ['4.90–50', '50–200', '200+']
df_cleaned['price_bin'] = pd.cut(df_cleaned['price'], bins=price_bins, labels=price_labels, include_lowest=True)

# Define custom rating bins and enforce desired order (4.6–4.9 on top)
rating_bins = [4.2, 4.5, 4.9]
rating_labels = ['4.2–4.5', '4.6–4.9']
df_cleaned['rating_bin'] = pd.cut(df_cleaned['rating'], bins=rating_bins, labels=rating_labels, include_lowest=True)

# Convert rating_bin to a categorical type with custom order
df_cleaned['rating_bin'] = pd.Categorical(df_cleaned['rating_bin'], categories=['4.6–4.9', '4.2–4.5'], ordered=True)

# Create pivot table
heatmap_data = df_cleaned.pivot_table(
    values='num_reviews',
    index='rating_bin',
    columns='price_bin',
    aggfunc='mean'  # or 'sum' if desired
)

# Plot heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap='YlOrRd', annot=True, fmt=".0f", linewidths=0.5)

plt.title('Heatmap of Average Number of Reviews by Rating and Price Range')
plt.xlabel('Price Range (€)')
plt.ylabel('Rating Range')
plt.tight_layout()
plt.show()
```

```python
# Define the price ranges and required counts
price_ranges = [(20, 50, 1), (50, 100, 2), (100, 150, 3), (150, 250, 4), (250, float('inf'), 5)]

# Create an empty list to store selected wines
selected_wines = []

# Loop through each price range and select the top wines
for min_price, max_price, count in price_ranges:
    wines_in_range = df_cleaned[(df_cleaned['rating'] > 4.5) & (df_cleaned['price'].between(min_price, max_price))]
    top_wines = wines_in_range.nlargest(count, 'num_reviews')
    selected_wines.append(top_wines)

# Concatenate all selections into a single dataframe
final_selection = pd.concat(selected_wines)[['wine', 'rating', 'num_reviews', 'price']]

# Remove duplicates, keeping only the one with the highest number of reviews per wine name
final_selection = final_selection.sort_values('num_reviews', ascending=False).drop_duplicates(subset='wine', keep='first')

# Ensure that exactly 5 wines are selected while keeping the price range logic
final_list = pd.DataFrame()
for min_price, max_price, _ in price_ranges:
    wine_in_range = final_selection[final_selection['price'].between(min_price, max_price)]
    if not wine_in_range.empty:
        final_list = pd.concat([final_list, wine_in_range.head(1)])  # Pick the highest-reviewed wine in the range

# Display the result
print(final_list)
```

```python
# Define price ranges and labels
price_ranges = [(20, 50), (50, 100), (100, 150), (150, 250), (250, float('inf'))]
price_labels = ['€20-50', '€50-100', '€100-150', '€150-250', '€250+']

# Calculate average number of reviews for each price range
avg_reviews = []
selected_wine_reviews = []
for (min_price, max_price), label in zip(price_ranges, price_labels):
    subset = df_cleaned[df_cleaned['price'].between(min_price, max_price)]
    avg_reviews.append(subset['num_reviews'].mean())

    # Find selected wine in this range
    selected_wine = final_list[final_list['price'].between(min_price, max_price)]
    if not selected_wine.empty:
        selected_wine_reviews.append(selected_wine.iloc[0]['num_reviews'])
    else:
        selected_wine_reviews.append(0)  # Assign 0 if no wine selected in this range

# Convert to DataFrame
comparison_df = pd.DataFrame({'Selected Wine Reviews': selected_wine_reviews, 'Avg Num Reviews': avg_reviews}, index=price_labels)

# Define bar width
bar_width = 0.4
x_labels = np.arange(len(price_labels))

# Create figure
fig, ax = plt.subplots(figsize=(10, 6))

# Plot grouped bars
ax.bar(x_labels - bar_width/2, comparison_df['Selected Wine Reviews'], bar_width, color='green', label='Wine Reviews')
ax.bar(x_labels + bar_width/2, comparison_df['Avg Num Reviews'], bar_width, color='darkgreen', label='Avg Reviews')

ax.set_ylabel('Number of Reviews')
ax.set_xticks(x_labels)
ax.set_xticklabels(price_labels, rotation=45, ha='right')

# Add legend
ax.legend(loc='upper left')

# Improve layout
plt.title('Comparison of Selected Wine Reviews vs Price Range Average')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```
