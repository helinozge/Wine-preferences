```python
import pandas as pd
df = pd.read_csv("wines_SPA.CSV")

df
```

```python

# Identify duplicated rows
duplicates = df.duplicated()

# Display duplicated rows
print("\nDuplicated Rows (identifies duplicates beyond the first occurrence):")
print(df[duplicates])

# Count total number of duplicate rows
total_duplicates = duplicates.sum()
print(f"\nTotal number of duplicate rows: {total_duplicates}")
```

```python
df_cleaned = df.drop_duplicates()

# Display the cleaned DataFrame
print("\nDataFrame after dropping duplicates (keeping first occurrence):")
print(df_cleaned)
```

```python

# Convert 'year' column to numeric, coercing errors to NaN
df_cleaned['year'] = pd.to_numeric(df_cleaned['year'], errors='coerce')

# Calculate the mean of the numeric 'year' column
year_mean = df_cleaned['year'].mean()

# Fill NaN values in the 'year' column with the mean
df_cleaned['year'] = df_cleaned['year'].fillna(year_mean)

# Display the DataFrame after filling NaN values
print("\nDataFrame after filling NaN values in 'year' with the mean:")
print(df_cleaned)
```
